```python
import math
top = hou.node("/obj")
#____________________________________________generate a colorpatch_______________________________________________________
def Gencolorpatch(r, g, b):
    containergrids = top.createNode("geo","squarecolor")
    
    
    square = containergrids.createNode("grid","type")
    square.parm("tx").set(brake*3)
    square.parm("orient").set("xy")
    square.parm("rows").set(2)
    square.parm("cols").set(2)
    square.parm("sizex").set(2)
    square.parm("sizey").set(2)
    square.parm("ty").set(2)
    squarecolor = containergrids.createNode("color","truecolor")
    squarecolor.parm("colorr").set(r)
    squarecolor.parm("colorg").set(g)
    squarecolor.parm("colorb").set(b)
    squarecolor.setInput(0, square)
    squarecolor.setDisplayFlag(True)
#______________________________________________generatefonts___________________________________________________________
def Genfonts(name):

    containerfonts = top.createNode("geo","fonts")
    fonts = containerfonts.createNode("font","type")
    fonts.parm("text").set(name)
    fonts.parm("fontsize").set(0.2)
    fonts.parm("ty").set(0.85)
    fonts.parm("tx").set(brake*3)
    fontscolor = containerfonts.createNode("color","black")
    fontscolor.parm("colorr").set(0)
    fontscolor.parm("colorg").set(0)
    fontscolor.parm("colorb").set(0)
    fontscolor.setInput(0, fonts)
    fontscolor.setDisplayFlag(True)
#______________________________________________colorcalculator___________________________________________________________
class color:
   #constructor
    def __init__(self, name="default", rval=0.0, gval=0.0, bval=0.0):
        '''this is my constructor for my class'''
        self.name = name
        self.rval = rval
        self.gval = gval
        self.bval = bval
    def __str__(self):
        '''string override for vector class'''
        return "color{} = ({},{},{})".format(self.name, self.rval, self.gval, self.bval)#, self.computeLength)
   
    def __setattr__(self, name, value):
       '''this my setattr override where I check the propper varibel name'''
       if name == "name" or name == "rval" or name == "gval" or name == "bval":
           self.__dict__[name] = value
       else:
           raise valueError, "setatter: invalid vector arrtribute name{}".format(name)
       
    def __add__(self,other):
        '''add my vector withr another vectore and return a new vector'''
        newVectorName = "{}+{}".format(self.name, other.name)
        rsum = self.rval + other.rval
        gsum = self.gval + other.gval
        bsum = self.bval + other.bval
        
        if rsum> 1.0:
            rsum = 1.0 
        if gsum > 1.0:
            gsum = 1.0
        if bsum > 1.0:
            bsum = 1.0
        
        return color(newVectorName, rsum, gsum, bsum)
      
    def __sub__(self,other):
        '''add my vector withr another vectore and return a new vector'''
        newVectorName = "{}-{}".format(self.name, other.name)
        rsub = self.rval - other.rval
        gsub = self.gval - other.gval
        bsub = self.bval - other.bval
        
        if rsub< 0.0:
            rsub = 0.0 
        if gsub < 0.0:
            gsub = 0.0
        if bsub < 0.0:
            bsub = 0.0
        return color(newVectorName, rsub, gsub, bsub)   
    
    def __invert__(self):
        ''' compute the inversion of my verctor to reutrn it as a new vector'''
        newVectorName = "~{}".format(self.name)
        return color(newVectorName, 1-self.rval, 1-self.gval, 1-self.bval)
    def __mul__(self,other):
        '''pberrode tje p[eratpr wotje vectpre crpss[rpdict amd retirem result'''
        newVectorName = "{}*{}".format(self.name, other.name)
        return color(newVectorName, self.rval*other.rval, self.gval*other.gval, self.bval*other.bval)
#_______________________________________enter some variable________________________________________________________
brake = 7 
containergrids = top.createNode("geo","name", run_init_scripts=False)
containerfonts = top.createNode("geo","name", run_init_scripts=False)
squarecolor = containergrids.createNode("color","truecolor")
A = color("A", .23, .45, .88)
B = color("B", 0.87, .31, .02)
C = A + B
D = A - B
E = ~A
F = ~B
G = ~A*~B
r = 0.0
g = 0.0
b = 0.0
#_____________________________use while loop to create 7 showcases_____________________________________
while brake > 0:
    
    if brake == 7:
        r = G.rval
        g = G.gval
        b = G.bval
        Genfonts("{}".format(G))
    elif brake == 6:
        r = F.rval
        g = F.gval
        b = F.bval
        Genfonts("{}".format(F))
    elif brake == 5:
        r = E.rval
        g = E.gval
        b = E.bval
        Genfonts("{}".format(E))
    elif brake == 4:
        r = D.rval
        g = D.gval
        b = D.bval
        Genfonts("{}".format(D))
    elif brake == 3:
        r = C.rval
        g = C.gval
        b = C.bval
        Genfonts("{}".format(C))
    elif brake == 2:
        r = B.rval
        g = B.gval
        b = B.bval
        Genfonts("{}".format(B))
        
    else: 
        r = A.rval
        g = A.gval
        b = A.bval
        Genfonts("{}".format(A))

    Gencolorpatch(r, g, b)
    brake = brake - 1 


print A
print B
print C
print D
print E
print F
print G
```
