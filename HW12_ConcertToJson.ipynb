```python
class jGeoDict():


    def __init__(self, inFile):
        '''Constructor for ObjectParent class'''
        self.inFile = open("TA14.txt", "r")
    
    def geometry(self):
        top = hou.node("/obj")
        container = top.createNode("geo","box_object1")
        geometry = container.createNode("box","box1")
    
    def translateGeoToDict(self, node):
    
        """ Given a Houdini node, get its points and primitive and pass them back as a dictionary"""
    
        geometry = node.geometry()                  # Get The Houdini Geometry
        tempDict = {}                                           # Create a temp Dictionary
        tempDict = self.parsePoints(geometry, tempDict)         # Get the points
        tempDict = self.parsePrimitives(geometry, tempDict)     # Get the primitives
        return tempDict
    
    def parsePoints(self, geometry, tempDict):
    
        """ From the geometry, grab the point information. For every point in the geometry create a mini-Dictionary containing the X, Y and Z coordiniates for that point"""
        points = geometry.points()                  # get the points from the geometry object
        pointList = []                              # create an empty list to contain each point dictionary
        
        # Go through all of the points and create a dictionary for each containing the points'
        # X, Y and Z values. Append that dictionary to the point list
        
        for point in points:
    
            pointDict = {}                          # create the empty Dictionary for the point
    
            pointDict['X'] = point.position()[0]    # Add the X Value
            pointDict['Y'] = point.position()[1]    # Add the Y value
            pointDict['Z'] = point.position()[2]    # Add the Z Value
            # Append the new dictionary to the point list
            pointList.append(pointDict)
    
        # Append the point list to the input Dictionary
        tempDict["points"] = pointList
    
        return tempDict
    
       
    
    def parsePrimitives(self, geometry, tempDict):
    
        """ From the geometry, grab the primitive information. For every primtive create a
            mini-Dictionary containing a smaller dictionary of all of its vertices"""
        prims = geometry.prims()
    
        primList = []
    
        # Go through every primitive and create a mini dictionary containing a list of all of its
        # vertices then append the mini dictionary to the primitive list
    
        for prim in prims:          # for every primitive
            primDict = {}           # create a mini Dictionary
            vcount = 0              # start the vertex count for each primitive at 0
    
            verts = prim.vertices() # Get the primitive's vertices
            for vert in verts:      # iterate through the vertices
    
                point = vert.point()    # get the point information
                primDict[str(vcount)] = point.number()  # Append each vertex point number to the Dict
                vcount += 1         # increment the vertex count in the primitive
    
            primList.append(primDict)   # Append the primtive Dictionary to the primList
        
        # add all of the primitives to the dictionary
        tempDict['primitives'] = primList
    
        return tempDict
```
