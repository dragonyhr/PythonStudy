```python
import random

class ObjectParent:
    '''create a base class to store fundamental information'''
    def __init__(self,name = 'default'):
        '''Constructor for ObjectParent class'''
        self.name = name

    def __str__(self):
        '''string override for the ObjectParent class'''
        return "Parent Path=/obj"
    
    def createContainer(self):
        '''create classs top mode and container for my obejects'''
        topnode = hou.node('/obj')
        self.container = topnode.createNode("geo", self.name)

class Geometry(ObjectParent):
    '''create a base class to store fundamental information need to create child class'''
    def __init__(self,name = 'default'):
        '''Constructor for Geometry class'''
        self.name = name
        self.createContainer()
        self.createGeo()
        self.color = (random.random(), random.random(), random.random())
        self.t = (random.uniform(-1, 1), random.uniform(-1,1), random.uniform(-1,1))
        self.r = (random.uniform(-90, 90), random.uniform(-90, 90), random.uniform(-90, 90))
        self.s = (random.uniform(.5,1.0), random.uniform(.5,1.0), random.uniform(.5,1.0))
    
    def __str__(self):
        '''string override for the Geometry class'''
        return "Parent Path=/obj, name = {}".format(self.name)
    def __setattr__(self, name ,value):
        self.__setBasicAttr__(name, value)

    
    def __setBasicAttr__(self, name, value):
        '''if the attribute is color,traslation, rotation,and size then make modification here'''
        self.__dict__[name] = value
        if name == 'color':
            self.objColor.parmTuple('color').set(value)
        elif name == 't':
            self.container.parmTuple('t').set(value)
        elif name == 'r':
            self.container.parmTuple('r').set(value)
        elif name == 's':
            self.container.parmTuple('s').set(value)
            
    def createGeo(self):
        '''set the Geocolor to the container'''
        outName = '{}_outputNode'.format(self.name)
        self.objColor = self.container.createNode("color",'objectcolor')
    
    def GeoDisplay(self, object):
        '''set Geo display here'''
        self.objColor.setInput(0, object)       
        self.objColor.setDisplayFlag(True)


class Donut(Geometry):
    '''this is a child of geo'''
    def __init__(self, name='Donut', xRadius = 1.0, yRadius = .5):
        Geometry.__init__(self, name)
        self.createDonut()
        
        self.xRadius =xRadius
        self.yRadius =yRadius
    
    
    def __str__(self):
        
        '''string override for Donut class'''
        return "name = {}, outerRadius = {},innerRadius = {}".format(self.name, self.xRadius, self.yRadius)
    
    def createDonut(self):
        
        self.Donut = self.container.createNode("torus", "Donut")
        self.GeoDisplay(self.Donut)
    
    def __setattr__(self, name ,value):
        
        self.__setBasicAttr__(name, value)
        if name == 'xRadius':
            self.Donut.parm('radx').set(value)
        if name == 'yRadius':
            self.Donut.parm('rady').set(value)
    

class Alberto(Geometry):
    '''this is a child of geo'''
    def __init__(self, name='Box'):
        Geometry.__init__(self, name)
        self.createAlberto()
    
    def createAlberto(self):
        
        self.Alberto = self.container.createNode("box", "Box")
        self.GeoDisplay(self.Alberto)
        

object = Donut("stu", .5+.5*random.random(), .5*random.random())    


hou.node("/obj").layoutChildren()   
print object

print Geometry.__str__(object)

print ObjectParent.__str__(object)
#print ObjectParent.__str__(object)
```
