```python
#____________________access the houdini.obj and math
top = hou.node("/obj")
import math
import random

#____________________enter the dates to produce the objects____________________________

myInput = raw_input("enter a object type, scale, numbers of this objects and radius of circle:")
type, scale, num, radius = myInput.split()
scale = float(scale)
num = int(num)
radius = float(radius)
brake = num

#____________________check errors start here____________________________________________


canexecute = True
error = ""

if type == "tube" or type == "torus" or type == "sphere" or type == "box":
    print type, "is good"
else: 
    canexecute = False
    error += "sorry, there is no {} here. you can only choose 'tube' or 'torus' or 'sphere' or 'box'".format(type)   

if scale >= 0.01 and scale <= 10:
    print "scale",scale, "is good"
else:
    canexecute = False
    print "sorry, the scale have to between 0.01 to 10."
    
    
if num == int(num) and num > 0 and num < 100:
    print "number" ,num, "is good"
else:
    canexecute = False
    error = error, "sorry the numbers of objects have to between from 0 to 100 integer."


if radius > 0:
    print "radius" ,radius, "is good"
else: 
    canexecute = False
    print "sorry,the radius have to be postive."
    
#____________________check errors end here_____________________________________________
#____________________if everything is good,the program will execute from here__________

if canexecute == True:
    while brake > 0:
    
        mytreasures = top.createNode("geo","mytreasures")
        object = mytreasures.createNode(type,"wisechoice")
        objectcolor = mytreasures.createNode("color","truecolor")
        objectcolor.parm("colorr").set(random.random())
        objectcolor.parm("colorg").set(random.random())
        objectcolor.parm("colorb").set(random.random())
        objectcolor.setInput(0, object)
        objectcolor.setDisplayFlag(True)
        #objectcolor.parm("colortype").set("Constant")
        
        object.parm("scale").set(scale)
        object.parm("tx").set(radius*math.cos(2*math.pi/num*brake))        
        object.parm("tz").set(radius*math.sin(2*math.pi/num*brake))
        brake = brake - 1
else:
    print error
    
#____________________The end of program________________________________________________




##access the houdini .obj and math
#top = hou.node("/obj")
#import math
#
##enter the dates to produce the objects
#
#myInput = raw_input("choose a object type, how many of this objects, scale and radius of circle:")
#type, num, scale, radius = myInput.split()
##print "type is:", type, "/number is :", num, "/scale = ", scale, "/radius = ", radius
#
#num = int(num)
#scale = float(scale)
#radius = float(radius)
#brake = num
#
#if type == "tube" or type == "torus" or type == "sphere" or type == "box":
#    if num == int(num) and num > 0 and num < 100:
#        if scale >= 0.01 and scale <= 10:
#            if radius > 0:
#           
#                while brake > 0:
#                
#                    mytreasures = top.createNode("geo","mytreasures")
#                    object = mytreasures.createNode(type,"wisechoice")
#                    object.setDisplayFlag(True)
#                    object.parm("scale").set(scale)
#                    object.parm("tx").set(radius*math.cos(2*math.pi/num*brake))
#                    object.parm("tz").set(radius*math.sin(2*math.pi/num*brake))
#                    brake = brake - 1
#            
#            else: 
#                print "sorry,the radius have to be postive."
#        else:
#            print "sorry, the scale have to between 0.01 to 10"
#    else:
#        print "sorry the numbers of objects have to between from 0 to 100 integer."
#else: 
#    print "sorry, there is no ", type,"here. you can only choose 'tube' or 'torus' or 'sphere' or 'box'"
     
#else: input = raw_input("sphere"):




#question_1 = raw_input("Which objects do you want: tube torus sphere or box")
#question_2 = raw_input("how many objects do you want?")
```
